```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
excel_path = 'Amazon_Germany_Sales_2022_to_2024.xlsx'  # Update if necessary
df = pd.read_excel(excel_path, sheet_name='Sheet1')

# Group by Category and sum Sales over all years
df_grouped = df.groupby('Category')['Sales_EUR_M'].sum().reset_index()

# Plot Pie Chart
plt.figure(figsize=(10, 8))
plt.pie(df_grouped['Sales_EUR_M'],
        labels=df_grouped['Category'],
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.Set3.colors,   # Nice readable color set
        wedgeprops={'edgecolor': 'white'})

plt.title('Total Amazon Germany Sales by Category (2022-2024)', fontsize=16, weight='bold')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

excel_path = 'Amazon_Germany_Sales_2022_to_2024.xlsx'
df = pd.read_excel(excel_path, sheet_name='Sheet1')

df_clean = df[['Category', 'Year', 'Sales_EUR_M']]

df_pivot = df_clean.pivot(index='Year', columns='Category', values='Sales_EUR_M')

plt.figure(figsize=(12, 7))

for category in df_pivot.columns:
    plt.plot(df_pivot.index, df_pivot[category], marker='o', label=category)

plt.title('Amazon Germany Sales Trend by Category (2022-2024)', fontsize=16, weight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Sales (€ Million)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

excel_path = 'Amazon_Germany_Sales_2022_to_2024.xlsx'  # Update path if necessary
df = pd.read_excel(excel_path, sheet_name='Sheet1')


df_grouped = df.groupby('Category').agg({
    'Avg_Price_EUR': 'mean',
    'Sales_EUR_M': 'sum'
}).reset_index()


plt.figure(figsize=(10, 7))

scatter = plt.scatter(df_grouped['Avg_Price_EUR'],
                      df_grouped['Sales_EUR_M'],
                      s=200,
                      c=range(len(df_grouped)),  # Color by category index
                      cmap='tab10',
                      alpha=0.8,
                      edgecolors='black')


for i, row in df_grouped.iterrows():
    plt.text(row['Avg_Price_EUR'] + 1, row['Sales_EUR_M'] + 50, row['Category'], fontsize=9)

plt.title('Average Price vs. Total Sales by Category (2022–2024)', fontsize=16, weight='bold')
plt.xlabel('Average Price (€)', fontsize=12)
plt.ylabel('Total Sales (€ Million)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
